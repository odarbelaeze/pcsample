```python
%matplotlib
import matplotlib.pyplot as plt

from pcsample import Sample

s = Sample(5, 4, 4)
r_latt = s.in_circles()
r_rand = s.out_circles()

plt.plot(r_latt[:,0], r_latt[:,1], '.r')
plt.plot(r_rand[:,0], r_rand[:,1], '.k')
#plt.show()
```
